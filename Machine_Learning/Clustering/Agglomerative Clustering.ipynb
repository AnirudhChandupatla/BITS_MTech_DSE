```python
def linkage(data,c1,c2, Type,distance):

    dists = []
    for p1 in c1:
        for p2 in c2:
            if(distance == 'euclidean'):
                td = np.linalg.norm(data[p1]-data[p2])
            else:
                td = sum(abs(value1 - value2) for value1, value2 in zip(data[p1], data[p2]))
            dists.append(td)
    dists = sorted(dists)
    if(Type == 'single'):
        dist = dists[0]
    if(Type == 'complete'):
        dist = dists[-1]
    if(Type == 'average'):
        dist = sum(dists)/len(dists)
            
    return round(dist,3)
```

```python
def build_distance_mat(data,clusters,link = 'single',distance='euclidean'):
    lc = len(clusters)
    distMat = [[0]*lc for i in range(lc)]
    for i in range(lc):
        for j in range(lc):
            if(i == 0 and j == 0):
                distMat[i][j] = 0
                continue
            distMat[i][j] = linkage(data,clusters[i],clusters[j],link,distance)
            print('D[{},{}] = {}'.format(clusters[i],clusters[j],distMat[i][j]))
    return distMat
    
```

```python
def get_min(mat,clusters):
    ll = len(mat)
    Min = float('inf')
    x,y = -1,-1
    for i in range(ll):
        for j in range(ll):
            if(mat[i][j] != 0 and mat[i][j] < Min):
                Min = mat[i][j]
                x,y = i,j
    print('clusters {} & {} have minimum distance of {}'.format(clusters[x],clusters[y],Min))
    return x,y
```

```python
def combine_clusters(clusters,x,y):
    new_combined_clusters = []
    for i in range(len(clusters)):
        if(i==y): continue
        if(i==x):
            new_combined_clusters.append(clusters[x]+clusters[y])
        else: new_combined_clusters.append(clusters[i])
    return new_combined_clusters
```

```python
import numpy as np
import pandas as pd
points = '2,2 5,8 2,4 4,3 3,5'
points = [list(map(float,p.split(','))) for p in points.split(' ')]
data = np.array(points)
data
```

```python
init_clusters = [[i] for i in range(len(data))]
init_clusters
```

```python

clusters = init_clusters
distance='euclidean'
while len(clusters) > 1:
    dist_mat = build_distance_mat(data,clusters,distance=distance)
    df = pd.DataFrame(data=dist_mat , columns=['-'.join([str(ee) for ee in e]) for e in clusters])
    print(df)
    x,y = get_min(dist_mat,clusters)
    print('since cluster {} & {} are near combining them'.format(clusters[x],clusters[y]))
    clusters = combine_clusters(clusters,x,y)
    print('new clusters are',clusters)
    print('-'*20,)
```
