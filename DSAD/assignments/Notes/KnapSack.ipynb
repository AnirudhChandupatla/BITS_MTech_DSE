## **Fractional**

```python
N = 7
MaxW = 15
O = list(range(1,N+1))
P = list(map(float,'5 10 15 7 8 9 4'.split(' ')))
W = list(map(float,'1 3 5 4 1 3 2'.split(' ')))
PW = [p/w for p,w in zip(P,W)]
```

```python
import pandas as pd
df = pd.DataFrame(data = [ (p,w,round(pw,3)) for p,w,pw in zip(P,W,PW)],columns=['Profit','Weight','P/W'])
df
```

```python
df = df.sort_values('P/W',ascending=False)
df
```

```python
w = MaxW
p = 0
i = 0
while w > 0 and i < N:
    cw = df.iloc[i].Weight    
    cp = df.iloc[i].Profit
    print('Object {}'.format(df.iloc[i].name),'Weight {}'.format(cw),'Profit {}'.format(cp),'remaining weight: {}'.format(w))
    if(cw <= w):
        p += cp
        print('adding {}'.format(df.iloc[i].name),'to knapsack;','remaining weight: {} - {} = {}'.format(w,cw,w-cw),'Profit: {}'.format(p))
        w -= cw
    else: 
        print("Object {} won't fit completely, only portion {}/{} will be added.".format(df.iloc[i].name,w,cw),'Final Profit: {}'.format(round(p + (cp * (w/cw)),3)))
        break
    i += 1
```

## **0/1 Knapsack:**

```python
s='''5 10
4 40
6 30
3 50'''
items = [tuple(map(int,e.split())) for e in s.split('\n')]
items
```

```python
import pandas as pd
df = pd.DataFrame(data=items,columns=['Weight','Profit'])
df
```

```python
N=len(items)
W=10
iW = [[None]*(W+1) for i in range(N+1)]

for i in range(N+1):
    iW[i][0] = 0
    
for i in range(W+1):
    iW[0][i] = 0
    
iW
```

```python
V = pd.DataFrame(data=iW,columns=range(W+1))
V
```

```python
def frac_knapsack(V,df,N,W):
    bb = df.Profit.values
    ww = df.Weight.values
    V = V.values
    for i in range(1,N+1):
        print('for i = {}'.format(i), 'b[{}] = {}'.format(i-1,bb[i-1]), 'w[{}] = {}'.format(i-1,ww[i-1]))
        for w in range(1,W+1):
            print('---for w = {}'.format(w))
            if(ww[i-1] <= w):# // item i can be part of the solution
                print('-----w[{}] = {} <= {}(w) '.format(i,ww[i-1],w))
                if(bb[i-1] + V[i-1,w-ww[i-1]] > V[i-1,w]):
                    print('-------b[{}] + V[{},{}] = {} + {} = {} > {}'.format(i,i-1,w-ww[i-1],bb[i-1],V[i-1,w-ww[i-1]],bb[i-1] + V[i-1,w-ww[i-1]],V[i-1,w]))
                    print('-------V[{},{}] = b[{}] + V[{}, {}] = {} + {} = {}'.format(i,w,i-1,i-1,w - ww[i-1],bb[i-1] , V[i-1,w - ww[i-1]],bb[i-1] + V[i-1,w - ww[i-1]]))
                    V[i,w] = bb[i-1] + V[i-1,w - ww[i-1]]
                else:
                    print('-------b[{}] + V[{},{}] = {} + {} = {} < {}'.format(i,i-1,w-ww[i-1],bb[i-1],V[i-1,w-ww[i-1]],bb[i-1] + V[i-1,w-ww[i-1]],V[i-1,w]))
                    print('-------V[{},{}] = V[{},{}] = {}'.format(i,w,i-1,w,V[i-1,w]))
                    V[i,w] = V[i-1,w]
            else:
                print('---w[{}] = {} > {}(w) '.format(i,ww[i-1],w))
                print('---V[{},{}] = V[{},{}] = {}'.format(i,w,i-1,w,V[i-1,w]))
                V[i,w] = V[i-1,w] #// wi > w
            print(V)
    return V
```

```python
v = frac_knapsack(V,df,N,W)
```

```python
i = N
k = W
w = df.Weight.values
p = df.Profit.values
tp = 0
while i>0:
    if(v[i,k] != v[i-1,k]):
        print('Items {} {} contributed profit of {} by occupying weight {}'.format(i,df.iloc[i-1].values,p[i-1],w[i-1]))
        k = k-w[i-1]
        tp += p[i-1]
        i -= 1
    else: i-=1
print('total maximum profit is',tp)
```
