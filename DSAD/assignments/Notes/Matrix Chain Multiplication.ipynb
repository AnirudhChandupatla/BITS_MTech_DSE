```python
def C(i,j,CM , d):
    MinC = float('inf')
    MinK = None
    l=[]
    for k in range(i,j):
        print( CM[i][k] , CM[k+1][j] , (d[i] , d[k+1] , d[j+1]))
        try:
            c = CM[i][k] + CM[k+1][j] + (d[i] * d[k+1] * d[j+1])
        except:
            #print(i,j,k)
            pass
        l.append(c)
        if(c<MinC):
            MinC = c
            MinK = k
        print('C[{},{}] + C[{},{}] + (d[{}] x d[{}] x d[{}]) = {} + {} + ({} x {} x {}) = '.format(i,k,k+1,j,i,k+1,j+1,CM[i][k],CM[k+1][j],d[i],d[k+1],d[j+1]),c)
    K_Mat[i][j] = MinK+1
    print('C[{},{}] = min({})'.format(i,j,l))
    return MinC
```

```python
def get_pair_with_diff(diff,n):
    l = []
    for i in range(n):
        for j in range(i,n):
            if(j-i == diff):
                l.append((i,j))
    return l
```

```python
from pprint import pprint
import pandas as pd
```

```python
def mat_chain_mul(d,n,Cost_Mat,mat_dims):
    for diff in range(1,n):
        for i,j in get_pair_with_diff(diff,n):
            print('--for [{},{}]'.format(i,j))
            Cost_Mat[i][j] = C(i,j,Cost_Mat,mat_dims)
            print('K matrix')
            pprint(pd.DataFrame(data=K_Mat,columns=range(n)))
            print('Cost matrix')
            pprint(pd.DataFrame(data=Cost_Mat,columns=range(n)))
```

```python
n = 5

Cost_Mat = [[None]*n for i in range(n)]
K_Mat = [[None]*n for i in range(n)]

for i in range(n):
    Cost_Mat[i][i] = 0
    K_Mat[i][i] = 0
    
mat_dims = '4,10,3,12,20,7'
mat_dims = list(map(int,mat_dims.split(',')))
    
mat_chain_mul(mat_dims,n,Cost_Mat,mat_dims)
```
